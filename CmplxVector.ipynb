```python
import numpy as np

# Adición de vectores complejos
vector1 = np.array([complex(input("Ingrese la parte real del vector 1: "), input("Ingrese la parte imaginaria del vector 1: "))])
vector2 = np.array([complex(input("Ingrese la parte real del vector 2: "), input("Ingrese la parte imaginaria del vector 2: "))])
suma_vectores = np.add(vector1, vector2)
print("Suma de vectores complejos:", suma_vectores)

# Inverso (aditivo) de un vector complejo
vector = np.array([complex(input("Ingrese la parte real del vector: "), input("Ingrese la parte imaginaria del vector: "))])
inverso_vector = np.negative(vector)
print("Inverso del vector complejo:", inverso_vector)

# Multiplicación de un escalar por un vector complejo
escalar_vector = complex(input("Ingrese el escalar (parte real): "), input("Ingrese el escalar (parte imaginaria): "))
multiplicacion_escalar_vector = np.multiply(escalar_vector, vector)
print("Producto del escalar y el vector complejo:", multiplicacion_escalar_vector)

# Adición de matrices complejas
matriz1 = np.array([[complex(input("Ingrese la parte real de la matriz 1, fila 1: "), input("Ingrese la parte imaginaria de la matriz 1, fila 1: "))],
                   [complex(input("Ingrese la parte real de la matriz 1, fila 2: "), input("Ingrese la parte imaginaria de la matriz 1, fila 2: "))]])
matriz2 = np.array([[complex(input("Ingrese la parte real de la matriz 2, fila 1: "), input("Ingrese la parte imaginaria de la matriz 2, fila 1: "))],
                   [complex(input("Ingrese la parte real de la matriz 2, fila 2: "), input("Ingrese la parte imaginaria de la matriz 2, fila 2: "))]])
suma_matrices = np.add(matriz1, matriz2)
print("Suma de matrices complejas:", suma_matrices)

# Inversa (aditiva) de una matriz compleja
matriz = np.array([[complex(input("Ingrese la parte real de la matriz, fila 1: "), input("Ingrese la parte imaginaria de la matriz, fila 1: "))],
                   [complex(input("Ingrese la parte real de la matriz, fila 2: "), input("Ingrese la parte imaginaria de la matriz, fila 2: "))]])
inverso_matriz = np.negative(matriz)
print("Inversa de la matriz compleja:", inverso_matriz)

# Multiplicación de un escalar por una matriz compleja
escalar_matriz = complex(input("Ingrese el escalar para la matriz (parte real): "), input("Ingrese el escalar para la matriz (parte imaginaria): "))
multiplicacion_escalar_matriz = np.multiply(escalar_matriz, matriz)
print("Producto del escalar y la matriz compleja:", multiplicacion_escalar_matriz)

# Transpuesta de una matriz/vector
transpuesta_matriz = np.transpose(matriz)
print("Transpuesta de la matriz/vector:", transpuesta_matriz)

# Conjugada de una matriz/vector
conjugada_matriz = np.conj(matriz)
print("Conjugada de la matriz/vector:", conjugada_matriz)

# Adjunta (daga) de una matriz/vector
adjunta_matriz = np.conj(np.transpose(matriz))
print("Adjunta de la matriz/vector:", adjunta_matriz)

# Producto de dos matrices (de tamaños compatibles)
def producto_matrices():
    matriz_a = np.array([complex(input("Ingrese la parte real e imaginaria de la primera matriz, separadas por espacio: ")) for _ in range(int(input("Ingrese el número de filas de la primera matriz: ")))]).reshape(-1, int(input("Ingrese el número de columnas de la primera matriz: ")))
    matriz_b = np.array([complex(input("Ingrese la parte real e imaginaria de la segunda matriz, separadas por espacio: ")) for _ in range(int(input("Ingrese el número de filas de la segunda matriz: ")))]).reshape(-1, int(input("Ingrese el número de columnas de la segunda matriz: ")))
    return np.dot(matriz_a, matriz_b)

# Función para calcular la "acción" de una matriz sobre un vector
def accion_matriz_vector():
    matriz_accion = np.array([complex(input("Ingrese la parte real e imaginaria de la matriz, separadas por espacio: ")) for _ in range(int(input("Ingrese el número de filas de la matriz: ")))]).reshape(-1, int(input("Ingrese el número de columnas de la matriz: ")))
    vector_accion = np.array([complex(input("Ingrese la parte real e imaginaria del vector, separadas por espacio: "))]).reshape(-1, 1)
    return np.dot(matriz_accion, vector_accion)

# Producto interno de dos vectores
def producto_interno():
    vector_interno1 = np.array([complex(input("Ingrese la parte real e imaginaria del primer vector, separadas por espacio: "))]).reshape(-1, 1)
    vector_interno2 = np.array([complex(input("Ingrese la parte real e imaginaria del segundo vector, separadas por espacio: "))]).reshape(-1, 1)
    return np.dot(np.conj(vector_interno1).T, vector_interno2)

# Norma de un vector
def norma_vector():
    vector_norma = np.array([complex(input("Ingrese la parte real e imaginaria del vector, separadas por espacio: "))]).reshape(-1, 1)
    return np.linalg.norm(vector_norma)

# Distancia entre dos vectores
def distancia_vectores():
    vector_distancia1 = np.array([complex(input("Ingrese la parte real e imaginaria del primer vector, separadas por espacio: "))]).reshape(-1, 1)
    vector_distancia2 = np.array([complex(input("Ingrese la parte real e imaginaria del segundo vector, separadas por espacio: "))]).reshape(-1, 1)
    return np.linalg.norm(vector_distancia1 - vector_distancia2)

# Valores y vectores propios de una matriz
def valores_vectores_propios():
    matriz_eigen = np.array([complex(input("Ingrese la parte real e imaginaria de la matriz, separadas por espacio: ")) for _ in range(int(input("Ingrese el tamaño de la matriz (número de filas): ")))]).reshape(-1, int(input("Ingrese el número de columnas de la matriz: ")))
    return np.linalg.eig(matriz_eigen)

# Revisar si una matriz es unitaria
def es_unitaria():
    matriz_unitaria = np.array([complex(input("Ingrese la parte real e imaginaria de la matriz, separadas por espacio: ")) for _ in range(int(input("Ingrese el tamaño de la matriz (número de filas): ")))]).reshape(-1, int(input("Ingrese el número de columnas de la matriz: ")))
    return np.allclose(np.dot(matriz_unitaria, np.conj(matriz_unitaria.T)), np.eye(matriz_unitaria.shape[0]))

# Revisar si una matriz es Hermitiana
def es_hermitiana():
    matriz_hermitiana = np.array([complex(input("Ingrese la parte real e imaginaria de la matriz, separadas por espacio: ")) for _ in range(int(input("Ingrese el tamaño de la matriz (número de filas): ")))]).reshape(-1, int(input("Ingrese el número de columnas de la matriz: ")))
    return np.allclose(matriz_hermitiana, np.conj(matriz_hermitiana.T))

# Producto tensor de dos matrices/vectores
def producto_tensor():
    matriz_tensor1 = np.array([complex(input("Ingrese la parte real e imaginaria de la primera matriz/vector, separadas por espacio: ")) for _ in range(int(input("Ingrese el tamaño de la primera matriz/vector (número de filas): ")))]).reshape(-1, int(input("Ingrese el número de columnas de la primera matriz/vector: ")))
    matriz_tensor2 = np.array([complex(input("Ingrese la parte real e imaginaria de la segunda matriz/vector, separadas por espacio: ")) for _ in range(int(input("Ingrese el tamaño de la segunda matriz/vector (número de filas): ")))]).reshape(-1, int(input("Ingrese el número de columnas de la segunda matriz/vector: ")))
    return np.kron(matriz_tensor1, matriz_tensor2)
```
